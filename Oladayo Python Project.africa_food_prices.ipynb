### RETAIL ANALYSIS

### The Analysis and Visualizations below is based on the Top ten Smallest prices for Retail Market Type. N.B- No comparative analysis and visualization can be made for the largest prices for retail market

```python
ret_mrt = dz[dz["market_type"] == "Retail"]
ret_mrt
```

```python
ret_mrt[["country","state","market","produce","quantity","month","year","price"]]
```

```python
# least Price
Tr = ret_mrt[["country","state","market","produce","quantity","month","year","price"]].nsmallest(n=10, columns= "price")
Tr
```

```python
plt.figure(figsize=(10,6))
plt.pie(Tr.price, autopct="%0.1f%%", normalize=True, labels = Tr.country) 
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.pie(Tr.month, autopct="%0.1f%%", normalize=True, labels = Tr.market) 
plt.show()
```

```python
x = Tr.market
y = Tr.price
colors = ("red","blue", "green", "yellow", "pink","black")

plt.figure(figsize=(12,6))
plt.bar(x,y, color = colors)
plt.title("Market Prices")
plt.xlabel("Market")
plt.ylabel("Prices")
plt.xticks(rotation=90)
plt.show()
```

```python
x = Tr.produce
y = Tr.price
colors = ("red","blue", "green", "yellow", "pink","black")

plt.figure(figsize=(12,6))
plt.bar(x,y, color = colors)
plt.title("Product Prices")
plt.xlabel("Produce")
plt.ylabel("Prices")
plt.xticks(rotation=90)
plt.show()
```

```python
x = Tr.market
y = Tr.month
colors = ("red","blue", "green", "yellow", "pink","black")

plt.figure(figsize=(12,6))
plt.bar(x,y, color = colors)
plt.title("Count of month per market")
plt.xlabel("Market")
plt.ylabel("Month")
plt.xticks(rotation=90)
plt.show()
```
